# Trigonometric Functions and Identities

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

## Graph

```python
# domain data
x = np.linspace(-2*np.pi, 2 * np.pi, 1000)

# range data
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
# get rid of connecting lines between aymptotes
tan_x[:-1][np.abs(np.diff(tan_x)) > 1] = np.nan
csc_x = np.array([sym.csc(n) for n in x])
# get rid of connecting lines between aymptotes
csc_x[:-1][np.abs(np.diff(csc_x)) > 1] = np.nan
sec_x = np.array([sym.sec(n) for n in x])
# get rid of connecting lines between aymptotes
sec_x[:-1][np.abs(np.diff(sec_x)) > 1] = np.nan
cot_x = np.array([sym.cot(n) for n in x])
# get rid of connecting lines between aymptotes
cot_x[:-1][np.abs(np.diff(cot_x)) > 1] = np.nan

# axis and tick marks
plt.figure(figsize=(12, 8))
plt.axvline(0, color='darkgray', linewidth=3)
plt.axhline(0, color='darkgray', linewidth=3)
points = np.array([i * sym.pi/4 for i in range(-8, 9)])
labels = ['$' + sym.latex(i * sym.pi/4) + '$' for i in range(-8, 9)]
plt.xticks(points, labels)
plt.gca().set_ylim([-3, 3])

# plots
plt.plot(x, sin_x, color='red', label='$\sin{x}$')
plt.plot(x, cos_x, color='green', label='$\cos{x}$')
plt.plot(x, tan_x, color='blue', label='$\\tan{x}$')
plt.plot(x, csc_x, color='hotpink', label='$\csc{x}$')
plt.plot(x, sec_x, color='lime', label='$\sec{x}$')
plt.plot(x, cot_x, color='dodgerblue', label='$\cot{x}$')

plt.title('Trigonometric Functions')
plt.grid()
plt.legend()

plt.show()
```

## Trigonometric Functions in Terms of $\sin{x}$ and $\cos{x}$

$$
\begin{align}
\tan{x} &= \frac{\sin{x}}{\cos{x}} \\
\csc{x} &= \frac{1}{\sin{x}} \\
\sec{x} &= \frac{1}{\cos{x}} \\
\cot{x} &= \frac{1}{\tan{x}} = \frac{\cos{x}}{\sin{x}}
\end{align}
$$
