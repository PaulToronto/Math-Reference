# 3D Plotting in Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

## `projection='3d'`

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')
```

## 3D Points and Lines

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection = '3d')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')

ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 1, 2])
ax.set_zticks([0, 1, 2])

ax.plot3D(xs=np.array([0, 1, 2]),
          ys=np.array([0, 1, 2]),
          zs=np.array([0, 1, 2]), 
          c='dodgerblue')

#ax.scatter3D(xs=0, ys=0, zs=0, c='red')
#ax.scatter3D(xs=1, ys=1, zs=1, c='lime')
#ax.scatter3D(xs=2, ys=2, zs=2, c='blue')
ax.scatter3D(xs=np.array([0, 1, 2]),
          ys=np.array([0, 1, 2]),
          zs=np.array([0, 1, 2]), 
          c=['red', 'lime', 'blue'])

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# data for a 3D line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# data for 3D scatter points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')

plt.show()
```

## 3D Continuous line

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')

x = np.linspace(-4 * np.pi, 4 * np.pi, 50)
y = np.linspace(-4 * np.pi, 4 * np.pi, 50)
z = x**2 + y**2

ax.plot3D(x, y, z)

plt.show()
```

## 3D Contour Plot

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 50)
y = np.linspace(-4 * np.pi, 4 * np.pi, 50)
z = lambda x, y: x**2 + y**2

X, Y = np.meshgrid(x, y)
Z = z(X, Y)
```

```python
x.shape, y.shape, z(x, y).shape
```

```python
X.shape, Y.shape, Z.shape
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.contourf3D(X, Y, Z, levels=500, cmap='Greens')
#ax.view_init(60, 35)
plt.show()
```

The above plot is analogous to its 2D version

```python
contours = plt.contourf(X, Y, Z, levels=500, cmap='Greens')
plt.colorbar()
plt.show()
```

## Wireframes and Surface Plots

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.plot_wireframe(X, Y, Z)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                cmap='viridis', edgecolor='none')

plt.show()
```
