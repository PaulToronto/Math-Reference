# Matplotlib quiver

This notebook is about using `matplotlib.pyplot.quiver` to draw vectors.

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## `ax.add_line` and `plt.Line2D`

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.scatter(1, 1, s=30, color='navy')
ax.add_line(plt.Line2D(xdata=(1, 4),
                       ydata=(1, 3),
                       linewidth=3,
                       color='lime'))
ax.add_line(plt.Line2D(xdata=(0.3, 0.3), ydata=(0, 0.3), color='red'))
ax.add_line(plt.Line2D(xdata=(0, 0.3), ydata=(0.3, 0.3), color='navy'))

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xticks(np.linspace(-5, 5, 11))
plt.yticks(np.linspace(-5, 5, 11))
plt.grid()
plt.show()
```

## `plt.annotate`

- `xy` : (float, float)
    - The point *(x, y)* to annotate. The coordinate system is determined
    by *xycoords*.
- `xytext` : (float, float), default: *xy*
    - The position *(x, y)* to place the text at. The coordinate system
    is determined by *textcoords*.
- `xycoords`: default is 'data'
 - 'data': Use the coordinate system of the object being annotated

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.scatter(0, 0, s=30, color='navy')
plt.annotate(r'$\theta$', xy=(0, 0), xytext=(.5, .5), size=25)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xticks(np.linspace(-5, 5, 11))
plt.yticks(np.linspace(-5, 5, 11))
plt.grid()
plt.show()
```

## `plt.quiver`

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.scatter(0, 0, s=30, color='navy')

ax.quiver(0, 0, 1, 1, scale=1, units='xy')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xticks(np.linspace(-5, 5, 11))
plt.yticks(np.linspace(-5, 5, 11))
plt.grid()
plt.show()
```

## Putting it all together

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

# vector a
plt.quiver(0, 0, 5, 6, scale=1, units='xy',
           color='red', width=0.17)
plt.annotate('$\mathbf{a}$', xy=(2, 3.2),
             size=15, color='red')
# vector b
plt.quiver(0, 0, 9, 0, scale=1, units='xy',
           color='dodgerblue', width=0.17)
plt.annotate('$\mathbf{b}$', xy=(4.5, -0.7),
             size=15, color='dodgerblue')

# line L
x_l = np.linspace(-1, 10, 100)
y_l = np.zeros_like(x_l)
plt.plot(x_l, y_l, color='lime',
         linewidth=4, label='Line: $L$', zorder=-1)

# adjacent side
ax.add_line(plt.Line2D(xdata=(0, 5), ydata=(0, 0), color='black', linewidth=2))
plt.annotate('x', xy=(2.5, 0.2),
             size=14, color='black')
# opposite side
ax.add_line(plt.Line2D(xdata=(5, 5), ydata=(0, 6), color='black', linewidth=2))
plt.annotate('y', xy=(4.6, 3),
             size=14, color='black')

# right-angle
ax.add_line(plt.Line2D(xdata=(4.7, 4.7), ydata=(0, 0.3), color='black', linewidth=1))
ax.add_line(plt.Line2D(xdata=(4.7, 5), ydata=(0.3, 0.3), color='black', linewidth=1))

# theta
plt.annotate(r'$\theta$', xy=(0.8, 0.3), size=14)

plt.title('$proj_{b}a$')
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.xticks(np.linspace(-1, 10, 12))
plt.yticks(np.linspace(-1, 10, 12))
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

# vector a
plt.quiver(0, 0, 1, 8, scale=1, units='xy',
           color='red', width=0.17)
plt.annotate('$\mathbf{a}$', xy=(-0.1, 4),
             size=15, color='red')
# vector b
plt.quiver(0, 0, 3, 4, scale=1, units='xy',
           color='dodgerblue', width=0.17)
plt.annotate('$\mathbf{b}$', xy=(1.7, 1.7),
             size=15, color='dodgerblue')

# line L
x_l = np.linspace(-1, 10, 100)
y_l = (4/3) * x_l
plt.plot(x_l, y_l, color='lime',
         linewidth=4, label='Line: $L$', zorder=-1)

# adjacent side
ax.add_line(plt.Line2D(xdata=(21/5, 0), ydata=(28/5, 0), color='black', linewidth=2))
plt.annotate('x', xy=(3, 3),
             size=14, color='black')
# opposite side
ax.add_line(plt.Line2D(xdata=(21/5, 1), ydata=(28/5, 8), color='black', linewidth=2))
plt.annotate('y', xy=(2.7, 7),
             size=14, color='black')

# right-angle
ax.add_line(plt.Line2D(xdata=(99/25, 189/50), ydata=(289/50, 277/50), color='black', linewidth=1))
ax.add_line(plt.Line2D(xdata=(201/50, 189/50), ydata=(134/25, 277/50), color='black', linewidth=1))

# theta
plt.annotate(r'$\theta$', xy=(0.4, 1.3), size=14)

plt.title('$proj_{b}a$')
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.xticks(np.linspace(-1, 10, 12))
plt.yticks(np.linspace(-1, 10, 12))
plt.grid()
plt.legend(loc='upper left')
plt.show()
```
